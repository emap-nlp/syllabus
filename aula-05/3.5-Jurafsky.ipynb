Esse notebook é sobre o exercício 3.5 da terceira edição do "Speech and Language Processing" do Dan Jurafsky; possivelmente no link https://web.stanford.edu/~jurafsky/slp3/3.pdf. Esse exercício é sobre a necessidade do símbolo de fim da sentença.

Se nosso corpus fosse:  <br>
\<s\>ab <br>
\<s\>bb <br>
\<s\>ba <br>
\<s\>aa <br>
teriamos no nosso de modelo de bigramas as probabilidades:<br>
$P(b|{<}s{>}) = 2/4 = 1/2;  P(a|{<}s{>}) = 2/4 = 1/2.$<br>
$P(a|a) = 1/2; P(b|a) = 1/2.$<br>
$P(a|b) = 1/2; P(b|b) = 1/2.$<br>

Daí, as sentenças de tamanho 2 são $2^2 = 4$(todas do corpus) e cada uma delas tem probabilidade $1/2 \cdot 1/2 = 1/4$, totalizando probabilidade $1$ de ter uma sentença de tamanho 2. Olhando as de tamanho 3, temos $2^3 = 8$ casos (\<s\>aba, \<s\>abb, \<s\>bba, \<s\>bbb, \<s\>baa, \<s\>bab, \<s\>aaa, \<s\>aab) e cada um com probabilidade $1/2 \cdot 1/2 \cdot 1/2 = 1/8$, totalizando em 1 a probabilidade de uma sentença ter tamanho $3$; absurdo pois deveríamos ter a probabilidade de todas as sentenças somando 1.

Se as sentenças no treino tivessem o símbolo de fim de sentença, teríamos <br>
$P(b|{<}s{>}) = 2/4 = 1/2;  P(a|{<}s{>}) = 2/4 = 1/2. $<br>
$P(a|a) = 1/4; P(b|a) = 1/4; P({<}/s{>}|a) = 2/4 = 1/2. $<br>
$P(a|b) = 1/4; P(b|b) = 1/4; P({<}/s{>}|b) = 2/4 = 1/2.$<br>

Nesse caso, nas sequências de tamanho 2 teríamos $2^2 = 4$ casos todos com probabilidade $1/2 \cdot 1/4 \cdot 1/2 = 1/16$, somando probabilidade $1/4$; as de tamanho 3 são $2^3 = 8$ sequências cada uma com probabilidade $1/2 \cdot 1/4 \cdot 1/4 \cdot 1/2 = 1/64$, somando probabilidade $1/8$. Nesse caso, a sequência de tamanho $n \geq 1$ terá probabilidade $1/2^{2n}$ cada, como são $2^n$ sequências de tamanho $n$ a probabilidade de cada é $1/2^n$; a soma dessas probabilidades dá 1.
No caso de não ter símbolo terminal, é facil ver que sempre dá invalido. A probabilidade das sequências de tamanho 2 num alfabeto $X$ sempre é 
$$\sum_{x,y \in X} P(xy|{<}s{>}) = \sum_{x \in X} P(x|{<}s{>}) \left( \sum_{y \in X} P(y|x) \right) = \sum_{x \in X} P(x|{<}s{>}) = 1,$$
e esse raciocínio pode ser aplicado pra sequências de qualquer tamanho.
